```python
import math
nd = 'y'
while nd == 'y':
  try:
    n1 = int(input('n1: '))
    n2 = int(input('n2: '))
    op1 = input('연산자')
    op = ['+', '-', '*', '/', '**']
    if not op1 in op:
      raise Exception('잘못된 연산자')

    if op1 == '+':
      r = n1 + n2
    elif op1 == '-':
      r = n1 - n2
    elif op1 == '*':
      r = n1 * n2
    elif op1 == '/':
      if n2 == 0:
        raise ZeroDivisionError
      r = n1 / n2
    elif op1 == '**':
      r = math.pow(n1, n2)
    print('result: ', r)
    nd = input('더 할래요? y/n')
  except ValueError as ve:
    print(f'errpr: {ve}')
    print('잘못된 입력입니다. 정수를 입력해주세요')
    continue
  except ZeroDivisionError as e:
    print(f'errpr: {e}')
    print('잘못된 입력입니다. ')
    continue
  except Exception as e:
    print(f'errpr: {e}')
    continue



```
