```python
 
class Account:

    cnt_account = 0

    def __init__(self, name, money):
        self.name = name
        self.money = money
        self.bank = 'SC은행'
        self.account = '000-00-0000'
        self.cnt_deposit = 0
        self.deposit_history = []
        self.withdraw_history = []
     
        Account.cnt_account += 1 #Q2
        
    def get_account_num(self): #A3
        return Account.cnt_account
        

    def deposit(self, money): #Q4
        if money < 1:
            print('입금은 1원 이상 가능')
        else:    
            self.money += money
            
            self.deposit_history.append(money) #Q10
            self.cnt_deposit += 1
            print(f'{money}원 입금, 잔고: {self.money}원', self.cnt_deposit)
            if self.cnt_deposit == 5: #Q7
                self.money = self.money + (self.money * 0.001)
                self.cnt_deposit = 0
                print(f' 이자 입금 {self.money * 0.001}')

    def withdraw(self, money): #Q5
        if money > self.money:
            print('출금은 잔고 이하 가능')
        else:    
            self.money -= money
            print(f'{money} 출금')
           
            self.withdraw_history.append(money) #Q10

    def display_info(self):
        print(f'예금주: {self.name}, 은행: {self.bank}, 계좌번호: {self.account}, 잔고: {self.money}')


a = Account('a', 10)
a.deposit(100)
a.deposit(1000)
a.deposit(1000)
a.deposit(1000)
a.deposit(1000)
a.withdraw(100)
print('입금 내역:', a.deposit_history, 'ㅊ금 내역:', a.withdraw_history)
a.withdraw(100000)
b = Account('b', 1000000)
c = Account('c', 10000000)
account_list = [a, b, c] #Q8

print('계좌 갯수:', a.get_account_num())

for i in account_list:  #Q9
    if i.money >= 1000000:
        i.display_info()

            
```
