# 사각형 넓이 구하기

```python
### 사각형 넓이를 구하는 클래스 완성!
class Square:
    def __init__(self):
        try:
            self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))
        except ValueError as e:
            print('입력값을 확인하세요:', e)
        except Exception as e:
            print('Err:', e)  
        else:
            if self.square == 1:
                print('직사각형 함수는 rect()입니다.')
    
            elif self.square == 2:
                print('평행사변형 함수는 par()입니다.')
            
            elif self.square == 3:
                print('사다리꼴 함수는 trape()입니다.')
            
            else:
                print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        try:
            width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
            area = width * vertical
            result = '직사각형의 넓이는 : ' + str(area)
            return result
        except ValueError as e:
            print('입력값을 확인하세요:', e)
        except Exception as e:
            print('Err:', e)         
    # 평행사변형 메서드
    def par(self):
        try:
            w, h = map(int, input('가로, 높이를 입력하세요. 예시 : 가로,높이\n >>>').split(','))
            area = w * h
            result = '평행사변형의 넓이는 : ' + str(area)
            return result
        except ValueError as e:
            print('입력값을 확인하세요:', e)
        except Exception as e:
            print('Err:', e)    
       

    # 사다리꼴 메서드
    def trape(self):
        try:
            low, up, height = map(int, input('하변, 상변, 높이를 입력하세요: 하변, 상변, 높이\n >>>').split(','))
            area = ((low + up) * height) / 2
            result = '사다리꼴의 넓이는 : ' + str(area)
            return result
        except ValueError as e:
            print('입력값을 확인하세요:', e)
        except Exception as e:
            print('Err:', e)    


a = Square()  # 객체 생성 & 2, 3을 각각 입력해 봅시다.
```

```python
a.rect()
```

```python
# 평형사변형의 넓이
a.par()  # 3,2를 입력했을 때 넓이가 6이 나와야 합니다.
```

```python
# 사다리꼴의 넓이
a.trape()  # 3,2,1 을 입력했을 때 넓이가 2.5이 나와야 합니다.
```
